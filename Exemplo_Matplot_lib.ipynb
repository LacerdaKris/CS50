```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
//criando 200 amostras de 1 unidade com taxa de erro de 30
x, y = make_regression(n_samples = 200, n_features = 1, noise = 30)
```

```python
//mostrar plotagem da população inicial
plt.scatter(x, y)
plt.show()
```
